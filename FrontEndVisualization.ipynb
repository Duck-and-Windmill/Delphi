```python
import plotly.plotly as py
import plotly.graph_objs as go
from pymongo import MongoClient
client = MongoClient()
repo = client.repo
final_data  = repo["delphi.finalData"]

```

```python
act_type = ['Email', 'Call', 'Letter']
male_data = final_data.find({'sex':1})
female_data = final_data.find({'sex':0})
male_values = {
    "Email": 0,
    "Call": 0,
    "Letter": 0
}
female_values = {
    "Email": 0,
    "Call": 0,
    "Letter": 0
}
        
for person in male_data:
    male_values[person['activity_type']] += 1
    
for person in female_data:
    female_values[person['activity_type']] += 1

for key in male_values:
    male_values[key] = male_values[key]/male_data.count()
    
for key in female_values:
    female_values[key] = female_values[key]/female_data.count()
```

```python
trace1 = go.Bar(
    x="Email",
    y=male_values['Email'],
    name='Male',
    marker=dict(color=['rgb(0,0,200)'])
)
trace2 = go.Bar(
    x="Email",
    y=female_values['Email'],
    name='Female',
    marker=dict(color=['rgb(200,0,0)'])
)

trace3 = go.Bar(
    x="Call",
    y=male_values['Call'],
    name='Male',
    marker=dict(color=['rgb(0,0,200)']),
    showlegend=False
)
trace4 = go.Bar(
    x="Call",
    y=female_values['Call'],
    name='Female',
    marker=dict(color=['rgb(200,0,0)']),
    showlegend=False
)

trace5 = go.Bar(
    x="Letter",
    y=male_values['Letter'],
    name='Male',
    marker=dict(color=['rgb(0,0,200)']),
    showlegend=False
)
trace6 = go.Bar(
    x="Letter",
    y=female_values['Letter'],
    name='Female',
    marker=dict(color=['rgb(200,0,0)']),
    showlegend=False
)

data = [trace1, trace2, trace3, trace4, trace5, trace6]
layout = go.Layout(
    barmode='group',
    title = "% of Male & Females vs. Effective Means of Advertising",
    xaxis=dict(
        title='Types of Advertising' 
    ),
    yaxis=dict(
        title='% of Males & Females'
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```

```python
act_type = ['Email', 'Call', 'Letter']
single_data = final_data.find({'married':0})
married_data = final_data.find({'married':1})
single_values = {
    "Email": 0,
    "Call": 0,
    "Letter": 0
}
married_values = {
    "Email": 0,
    "Call": 0,
    "Letter": 0
}
        
for person in married_data:
    married_values[person['activity_type']] += 1
    
for person in single_data:
    single_values[person['activity_type']] += 1

for key in married_values:
    married_values[key] = married_values[key]/married_data.count()
    
for key in single_values:
    single_values[key] = single_values[key]/single_data.count()
```

```python
trace1 = go.Bar(
    x="Email",
    y=married_values['Email'],
    name='Married',
    marker=dict(color=['rgb(0,0,200)'])
)
trace2 = go.Bar(
    x="Email",
    y=single_values['Email'],
    name='Single',
    marker=dict(color=['rgb(200,0,0)'])
)

trace3 = go.Bar(
    x="Call",
    y=married_values['Call'],
    name='Married',
    marker=dict(color=['rgb(0,0,200)']),
    showlegend=False
)
trace4 = go.Bar(
    x="Call",
    y=single_values['Call'],
    name='Single',
    marker=dict(color=['rgb(200,0,0)']),
    showlegend=False
)

trace5 = go.Bar(
    x="Letter",
    y=married_values['Letter'],
    name='Married',
    marker=dict(color=['rgb(0,0,200)']),
    showlegend=False
)
trace6 = go.Bar(
    x="Letter",
    y=single_values['Letter'],
    name='Single',
    marker=dict(color=['rgb(200,0,0)']),
    showlegend=False
)

data = [trace1, trace2,trace3, trace4, trace5, trace6]
layout = go.Layout(
    barmode='group',
    title = "% of Married & Single Individuals vs. Effective Means of Advertising",
    xaxis=dict(
        title='Types of Advertising' 
    ),
    yaxis=dict(
        title='% of Married & Single Individuals'
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```
